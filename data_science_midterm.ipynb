# Analysis and Testing

Hypothesis: The number of deaths caused by covid-19 and exchange rates are positively correlated.
The hypothesis will be tested on 3 seperate countries, which are Japan, China and America respectively.

# Japan

## Correlation Check

```python
JPY_data.plot.scatter('JPY','deaths')
plt.title('JPY exchange rate vs number of deaths')
plt.xlabel('EUR/JPY')
plt.ylabel('log(deaths)')
```

```python
corr, p_value = pearsonr(JPY_data['JPY'], JPY_data['deaths'])
print("Pearson's Correlation Coefficient:", corr)
print("P-value:", p_value)
```

There is a strong positive correlation between the JPY and the number of covid-19 deaths

## Regress JPY exchange rates against log of covid-19 deaths

```python
mod = sm.OLS(JPY_data['JPY'], JPY_data['deaths'])
JPY_mod = mod.fit()
JPY_mod.summary()
```

The regression model shows there is a strong positive relationship between the number of deaths caused by covid-19 in Japan and the Japanese currency exchange rates. Since the P value is significant (<0.05) we could say that on average if log of covid-19 deaths increase by 1 unit, then the Chinese currency exchange rates would increase by 13.9 units. In other words, the Japanese currency would become weaker compared to Euro. In addtion, the adjusted R-squared statistics is very high (0.98). This suggest that the number of covid-19 deaths has a strong overall effect on exchange rate.

# Summary
1. The analysis produced mixed results.

2. While all the linear regression model suggest that there is a positive relationship between a country's currency and the number of covid-19 deaths, only Japan's Pearson's coefficient yields a positive relationship. By contrast, America and China's Pearson's coefficient suggest there is a negative relationship between the 2 variables.

3.  As all the R-squared are > 0.9, this suggest that linear regresson is a good model to capture the relationship between the number of covid-19 deaths and exchange rates. This makes it difficult to choose between the Pearson's coefficient and the linear regression coefficient.


# Conclusion:
